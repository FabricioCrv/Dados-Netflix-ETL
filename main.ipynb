```python
import pandas as pd
import os
import datetime as dt
```

```python
for dirname, _, filenames in os.walk('data'):
    for file in filenames:
        print(os.path.join(dirname, file))
```

```python
df = pd.DataFrame()

arquivoDevices = 'data/All_Devices.csv'
arquivoProfiles = 'data/All_Profiles.csv'
arquivoClickstream = 'data/All_Clickstream.csv'
arquivoSearchHistory = 'data/All_SearchHistory.csv'
arquivoViewingActivity = 'data/All_ViewingActivity.csv'
```

```python
#Lendo todos os dados em Profiles.csv
datalist = [pd.read_csv(arquivoProfiles)]
df = pd.concat(datalist)
df
```

```python
#Passando todos os "users" para uma chave, adicionando um suffixo do index para cada usuario
listaUsuarios = list(df['Profile Name'].value_counts().index)
usuariosDict = {profile: "User " + str(listaUsuarios.index(profile)) for profile in listaUsuarios}
df.replace({"Profile Name": usuariosDict}, inplace=True)
usuariosDict
```

```python
#Retirando as colunas desnecessárias
df.drop(columns=df.columns[1:], inplace=True)
df
```

```python
#Lendo os dados - Devices
datalist = [pd.read_csv(arquivoDevices)]
df = pd.concat(datalist)
df
```

```python
#inserindo o dicionário de usuários no dataframe
df.replace({"Profile Name": usuariosDict}, inplace=True)
df
```

```python
#Criando um dicionário para todos os ESNs 
listaEsn = list(df['Esn'].value_counts().index)
esnDict = {esn:"Esn " + str(listaEsn.index(esn)) for esn in listaEsn}
df.replace({"Esn": esnDict}, inplace=True)
esnDict
```

```python
#Criando um dicionário para os tipos de aparelho
listaTipoDevices = list(df['Device Type'].value_counts().index)
deviceTipoDict = {deviceType: "Device Type " + str(listaTipoDevices.index(deviceType)) for deviceType in listaTipoDevices}
df.replace({"Device Type" : deviceTipoDict}, inplace=True)
deviceTipoDict
```

```python
#Retirando as contas desativadas
df = df[df['Deactivation Time'].isnull()]
```

```python
#Criando novas colunas para a primeria e ultima data e hora de playback e retirando as demais
df['First Playback Date'] = pd.to_datetime(df["Acct First Playback Date"]).dt.date
df['Last Playback Date'] = pd.to_datetime(df["Acct Last Playback Date"]).dt.date

df['First Playback Time'] = pd.to_datetime(df["Acct First Playback Date"]).dt.time
df['Last Playback Time'] = pd.to_datetime(df["Acct Last Playback Date"]).dt.time

#Retirando os microsegundos das colunas de hora
df['First Playback Time'] = df['First Playback Time'].apply(lambda x: x.replace(microsecond=0))
df['Last Playback Time'] = df['Last Playback Time'].apply(lambda x: x.replace(microsecond=0))

df.drop(columns=['Acct First Playback Date',
       'Acct Last Playback Date',
       'Acct First Playback Date For User Generated Plays',
       'Acct Last Playback Date For User Generated Plays',
       'Profile First Playback Date', 'Profile Last Playback Date',
       'Profile First Playback Date For User Generated Plays',
       'Profile Last Playback Date For User Generated Plays',
       'Deactivation Time'], inplace=True)

df
```

```python
#Lendo os dados - Clickstream
datalist = [pd.read_csv(arquivoClickstream)]
df = pd.concat(datalist)
df
```

```python
#Colocando o dicionário de usuarios no dataframe
df.replace({"Profile Name": usuariosDict}, inplace=True)
```

```python
#Criando um dicionário para os Sources
listaSource = list(df['Source'].value_counts().index)
sourceDict = {source: "Source " + str(listaSource.index(source)) for source in listaSource}
df.replace({"Source": sourceDict}, inplace=True)
sourceDict
```

```python
#Criando novas colunas para a data e hora do Click e retirando as demais
df['Click Date'] = pd.to_datetime(df['Click Utc Ts']).dt.date
df['Click Time'] = pd.to_datetime(df['Click Utc Ts']).dt.time

df.drop(columns=['Referrer Url', 'Webpage Url', 'Click Utc Ts'], inplace=True)
df
```

```python
#Lendo os dados - Histórico
datalist = [pd.read_csv(arquivoSearchHistory)]
df = pd.concat(datalist)
df
```

```python
df.replace({"Profile Name": usuariosDict}, inplace=True)
```

```python
#Criando um dicionário os aparelhos utilizados
listaAparelhos = list(df['Device'].value_counts().index)
aparelhosDict = {devices:"Device " + str(listaAparelhos.index(devices)) for devices in listaAparelhos}
df.replace({"Devices": aparelhosDict}, inplace=True)
aparelhosDict
```

```python
df
```

```python
#Lendo os dados - Atividade de visualização
datalist = [pd.read_csv(arquivoViewingActivity)]
df = pd.concat(datalist)
df
```

```python
#inserindo os dicionários de usuarios e tipo de aparelho no dataframe
df.replace({"Profile Name": usuariosDict}, inplace=True)
df.replace({"Device Type" : deviceTipoDict}, inplace=True)
```

```python
#Criando novas colunas para data e hora em que o conteúdo foi visualizado, e retirando as colunas desnecessárias
df['Viewed at Date'] = pd.to_datetime(df['Start Time']).dt.date
df['Viewed at Time'] = pd.to_datetime(df['Start Time']).dt.time

df['Bookmarked in Minutes'] = pd.to_datetime(df['Bookmark']).dt.time

df.drop(columns=["Duration", "Bookmark", "Latest Bookmark", "Start Time", "Attributes", "Supplemental Video Type"], inplace=True)
df
```
